## Matplotlib labels

Plots need a description. What’s the use of a plot, if the viewer doesn’t know what the numbers represent.

```python
data = [23, 85, 72, 43, 52]
labels = ['A', 'B', 'C', 'D', 'E']
plt.xticks(range(len(data)), labels)
plt.xlabel('Class')
plt.ylabel('Amounts')
plt.title('Bar Chart')
plt.bar(range(len(data)), data)
plt.show()
```

## Multiple Charts

```python
data1 = [23, 85, 72, 43, 52]
data2 = [42, 35, 21, 16, 9]
width = 0.3
plt.bar(np.arange(len(data1)), data1, width=width)
plt.bar(np.arange(len(data2)) + width, data2, width=width)
plt.show()
```

## Stack Charts

```python
plt.bar(range(len(data1)), data1)
plt.bar(range(len(data2)), data2, bottom=data1)
plt.show()
```
